```
%%sql -r dataframe_24
SELECT
CUSTOMER_TYPE,
SUM(SALES_AMOUNT) AS TOTAL_SALES
FROM SALES_DATA
GROUP BY CUSTOMER_TYPE;
SELECT
SALES_CHANNEL,
SUM(SALES_AMOUNT) AS TOTAL_SALES
FROM SALES_DATA
GROUP BY SALES_CHANNEL;
SELECT
SALES_REP,
SUM(SALES_AMOUNT) AS TOTAL_SALES
FROM SALES_DATA
GROUP BY SALES_REP
ORDER BY TOTAL_SALES DESC
LIMIT 10;
SELECT
YEAR(SALE_DATE) AS YEAR,
MONTH(SALE_DATE) AS MONTH,
SUM(SALES_AMOUNT) AS TOTAL_SALES
FROM SALES_DATA
GROUP BY
YEAR(SALE_DATE),
MONTH(SALE_DATE)
ORDER BY
YEAR,
MONTH;
```

```
%%sql -r dataframe_25
SELECT
PRODUCT_CATEGORY,
AVG(DISCOUNT) AS AVERAGE_DISCOUNT
FROM SALES_DATA
GROUP BY PRODUCT_CATEGORY
ORDER BY AVERAGE_DISCOUNT DESC;
SELECT
PRODUCT_CATEGORY,
SUM(QUANTITY_SOLD) AS TOTAL_QUANTITY
FROM SALES_DATA
GROUP BY PRODUCT_CATEGORY
ORDER BY TOTAL_QUANTITY DESC;
SELECT
SALE_DATE,
PRODUCT_CATEGORY,
((UNIT_PRICE - UNIT_COST) * QUANTITY_SOLD) AS PROFIT
FROM SALES_DATA;
SELECT
REGION,
SUM((UNIT_PRICE - UNIT_COST) * QUANTITY_SOLD) AS TOTAL_PROFIT
FROM SALES_DATA
GROUP BY REGION
ORDER BY TOTAL_PROFIT DESC;
CREATE OR REPLACE VIEW SALES_SUMMARY AS
SELECT
REGION,
PRODUCT_CATEGORY,
SUM(SALES_AMOUNT) AS TOTAL_SALES,
SUM(QUANTITY_SOLD) AS TOTAL_QUANTITY
FROM SALES_DATA
GROUP BY
REGION,
PRODUCT_CATEGORY;
```

```
%%sql -r dataframe_26
SELECT *
FROM SALES_SUMMARY;
```
